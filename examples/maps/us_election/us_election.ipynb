```python
import requests

headers = {
    'accept': '*/*',
    'accept-language': 'en-GB,en;q=0.9',
    'origin': 'https://www.nytimes.com',
    'priority': 'u=1, i',
    'referer': 'https://www.nytimes.com/',
    'sec-ch-ua': '"Brave";v="129", "Not=A?Brand";v="8", "Chromium";v="129"',
    'sec-ch-ua-mobile': '?1',
    'sec-ch-ua-platform': '"Android"',
    'sec-fetch-dest': 'empty',
    'sec-fetch-mode': 'cors',
    'sec-fetch-site': 'cross-site',
    'sec-gpc': '1',
    'user-agent': 'Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/129.0.0.0 Mobile Safari/537.36',
}

response = requests.get(
    'https://static01.nyt.com/elections-assets/basemap-data/election2024/v6/2024/us/electoral-college-cartogram.topojson',
    headers=headers,
)
```

```python
electoral_map = response.json()

# Save TopoJson to file
with open('electoral-college-cartogram.json', 'w') as file:
    file.write(response.text)
```

---

Plot using Altair.

```python
import altair as alt
```

```python
electoral_map.keys()
```

```python
electoral_map['objects'].keys()
```

```python
electoral_map['objects']['innerlines'].keys()
```

```python
data = alt.Data(values=electoral_map, format=alt.DataFormat(feature='state_labels', type='topojson'))
chart = alt.Chart(data).mark_geoshape().project(
    type='identity',
    reflectY=True
)
chart.display()
```

```python
# Plot electoral map
data = alt.Data(values=electoral_map['objects']['innerlines']['geometries'])

alt.Chart(data).mark_geoshape()
```
