```python
import pandas as pd

df = nomis_df.copy()
df = df[["GEOGRAPHY_CODE", "MEASURES_NAME", "VARIABLE_NAME", "OBS_VALUE"]]
df = df.rename(columns={"GEOGRAPHY_CODE": "GSScode", "OBS_VALUE": "value", "MEASURES_NAME": "measure", "VARIABLE_NAME": "series"})
df = df.query("measure == 'Variable'")
df = df[['GSScode', 'series', 'value']]
df = df.pivot(index='GSScode', columns='series', values='value').reset_index(drop=False)

# Economic activity rate - aged 16-64    2528
# Employment rate - aged 16-64           2528
# % with RQF4+ - aged 16-64              2528

series_rebinds = {
    'Economic activity rate - aged 16-64': 'activity_rate',
    'Employment rate - aged 16-64': 'employment_rate',
    '% with RQF4+ - aged 16-64': 'rqf4_rate'
}
df = df.rename(columns=series_rebinds)

# add in the quintiles with range labels
for col in ['activity_rate', 'employment_rate', 'rqf4_rate']:
    quintiles, bins = pd.qcut(df[col], 5, retbins=True, labels=False, duplicates='drop')
    labels = [
        f"Lowest Quintile (<{bins[1]:.1f}%)",
        f"Second Quintile ({bins[1]:.1f}%-{bins[2]:.1f}%)",
        f"Third Quintile ({bins[2]:.1f}%-{bins[3]:.1f}%)",
        f"Fourth Quintile ({bins[3]:.1f}%-{bins[4]:.1f}%)",
        f"Highest Quintile ({bins[4]:.1f}%+)"
    ]
    df[f'{col}_quintile'] = pd.qcut(df[col], 5, labels=labels, duplicates='drop')

lfs_df = df.copy()
```

```python
wpc = gpd.read_file("uk-constituencies-2024-21-june-24.geojson")
wpc.set_crs("EPSG:27700", inplace=True)
wpc
```

```python
supermarkets = pd.read_csv("geolytix_retailpoints_v31_202403.csv")
supermarkets

# how many 'The Co-operative Food' supermarkets are there in each constituency?
supermarkets['brand'] = supermarkets['fascia'].str.lower()

bindings = json.loads('{  "the co-operative food": "co-operative",  "the co-operative food pfs": "co-operative",  "morrisons daily": "morrisons",  "morrisons": "morrisons",  "morrisons select": "morrisons",  "morrisons home and nutmeg": "morrisons",  "spar": null,  "spar pfs": null,  "heron": null,  "eurospar": null,  "tesco express": "tesco",  "tesco": "tesco",  "tesco express esso": "tesco",  "tesco extra": "tesco",  "marks and spencer": "marks and spencer",  "marks and spencer simply food": "marks and spencer",  "marks and spencer foodhall": "marks and spencer",  "marks and spencer bp": "marks and spencer",  "marks and spencer hospital": "marks and spencer",  "marks and spencer outlet": "marks and spencer",  "marks and spencer travel sf": "marks and spencer",  "marks and spencer msa": "marks and spencer",  "marks and spencer home": "marks and spencer",  "marks and spencer food to go": "marks and spencer",  "waitrose": "waitrose",  "little waitrose": "waitrose",  "little waitrose shell": "waitrose",  "waitrose msa": "waitrose",  "dunnes stores": null,  "iceland": "iceland",  "cook": null,  "lidl": "lidl",  "asda supermarket": "asda",  "asda express": "asda",  "asda on the move": "asda",  "asda superstore": "asda",  "asda supercentre": "asda",  "asda living": "asda",  "asda pfs": "asda",  "aldi": "aldi",  "aldi local": "aldi",  "sainsburys": "sainsburys",  "sainsburys local": "sainsburys",  "budgens": null,  "makro": null,  "farmfoods": null,  "the food warehouse": null,  "booths": null,  "costco": null,  "cooltrader": null,  "amazon fresh": null,  "planet organic": null,  "swift": null,  "whole foods market": null}')

supermarkets['brand'] = supermarkets['brand'].map(bindings)
supermarkets = supermarkets.dropna(subset=['brand'])

# get the number of supermarkets in each constituency w/ point in polygon
supermarkets = gpd.GeoDataFrame(supermarkets, geometry=gpd.points_from_xy(supermarkets.long_wgs, supermarkets.lat_wgs))
supermarkets = supermarkets[['brand', 'geometry']]
supermarkets = supermarkets.set_crs("EPSG:4326")
supermarkets = supermarkets.to_crs(wpc.crs)

joined = gpd.sjoin(supermarkets, wpc, how='inner', op='within')

# Group by GSScode and brand, then count the number of supermarkets in each constituency
supermarket_counts = joined.groupby(['GSScode', 'brand']).size().reset_index(name='count')
supermarket_counts = supermarket_counts.pivot(index='GSScode', columns='brand', values='count').reset_index(drop=False)
supermarket_counts

for brand in supermarket_counts.columns[3:].tolist():    
    supermarket_counts[f'has_{brand}'] = supermarket_counts[brand].notnull()
```

```python
for brand in supermarket_counts.columns[1:]:
    temp_df = supermarket_counts[['GSScode', brand]]
    temp_df = temp_df.merge(elec_df, on='GSScode', how='right')
    temp_df = temp_df.dropna(subset=[brand])
    number_con = temp_df.query("winner_2019 == 'Con'").shape[0]
    print(f"{brand} - {len(temp_df)} constituencies")
    print(f"Conservative constituencies: {number_con}")
    print(f"Con %: {number_con/len(temp_df)*100:.2f}%", end="\n\n")
```

```python
df = elec_df.copy()
df = df.merge(supermarket_counts[['GSScode', 'has_waitrose', 'has_lidl']], on='GSScode', how='left')
# Waitrose only, Morrisons only, Both, Neither
df['store_status'] = df.apply(lambda x: "Waitrose only" if x['has_waitrose'] and not x['has_lidl'] else "Morrisons only" if x['has_lidl'] and not x['has_waitrose'] else "Both" if x['has_waitrose'] and x['has_lidl'] else "Neither", axis=1)
df['Con_win'] = df['winner_2019'] == 'Con'
df['LD_win'] = df['winner_2019'] == 'LD'
df.groupby('store_status')['Con_win'].mean()
```
